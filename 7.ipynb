### （二）素材文件“销售数据.xlsx”中存放了某公司所有门店2021年前半年的销售数据，请根据此文件完成下面的操作。
#### 1.将“销售数据.xlsx”中的“销售数据”工作表数据读入DataFrame对象sales_data中。
#### 2.计算每笔销售业务的销售金额，并将计算结果新增到sales_data中（销售金额=销售数量*单价）。
#### 3.计算2021年前半年的销售金额合计。
#### 4.统计2021年前半年销售业务中，各产品的最大销售量。
#### 5.统计2021年前半年销售业务中，各门店各产品的销售金额合计。
#### 6.计算每个月份每种产品的总销量。
#### 7.对每个月的销售数量进行统计性描述输出。
#### 8.对每个门店的销售金额进行统计性描述输出。
#### 9.统计2021年前半年每种产品的销售频次。
#### 10.计算每个月的销售利润总和，并按升序排序计算结果。

```python
# 1.
import pandas as pd   # 导入pandas包
sales_data=pd.read_excel("销售数据.xlsx",sheet_name='销售数据')
sales_data
```

```python
# 2.
sales_data['销售金额'] = sales_data['销售数量'] * sales_data['单价']
sales_data
```

```python
# 3.
# 此数据集中的数据都是前半年的，1月-6月
sales_data['销售金额'].sum()
```

```python
# 4.
sales_data['销售数量'].groupby(sales_data['产品编码']).max()
```

```python
# 5.
temp = sales_data.groupby(['门店编码','产品编码'])
temp["销售金额"].sum()
```

```python
# 6.
temp = sales_data.groupby(['月份','产品编码'])
temp["销售数量"].sum()
```

```python
# 7.
sales_data["销售数量"].groupby(sales_data["月份"]).describe()
```

```python
# 8.
sales_data["销售金额"].groupby(sales_data["门店编码"]).describe()
```

```python
# 9.
sales_data["销售数量"].groupby(sales_data["产品编码"]).count()
```

```python
# 10.
sales_data["销售金额"].groupby(sales_data["月份"]).sum().sort_values()
```
