### 2.对读入的数据进行行列转置

### 3.根据下面的公式计算中国软件的销售毛利率和净资产收益率。
销售毛利率 =（销售净收入-产品成本）/销售净收入<p>
净资产收益率 = 净利润/((期初净资产+期末净资产)/2)


```python
import numpy as np
v1 = df1['销售净收入']- df1['产品成本']
v2 = np.divide(v1,df1['销售净收入'])  # 如果不需要把“销售毛利率”作为新的一列增加到原数据集的时候
v2
```

```python
df1["销售毛利率"] = np.divide(v1,df1['销售净收入'])  # 如果要把“销售毛利率”作为新的一列增加到原数据集的时候
df1
```

```python
v2 = (df1['期初净资产']+df1['期末净资产'])/2
df1['净资产收益率'] = np.divide(df1['净利润'],v2)  # 增加“净资产收益率”一列
df1
```

### 4. 绘制折线图,展现中国软件的销售毛利率和净资产收益率

```python
import matplotlib.pyplot as plt  # 导入绘图包

plt.rcParams['font.sans-serif']=['SimHei']      # 设置正常显示中文
plt.rcParams['axes.unicode_minus']=False     # 设置正常显示负号

fig1 = plt.figure(figsize=(20,4))        # 指定figure的宽和高，单位为英寸
plt.xticks(rotation=60)                # 设置x轴标签的旋转角度
plt.plot(df1.index,df1["销售毛利率"])    # 绘制"销售毛利率"的折线图
plt.plot(df1.index,df1["净资产收益率"])  # 绘制"净资产收益率"的折线图
plt.legend(["销售毛利率","净资产收益率"])  # 添加图例
plt.title('中国软件盈利能力分析')                  # 添加图表标题
plt.xlabel('季度')                     # 添加X轴标签
plt.ylabel('比率')                     # 添加y轴标签
```

## 方法二

```python
# 1.读取“中国软件资产负债表.xlsx”文件中的数据
import pandas as pd   
df = pd.read_excel("中国软件资产负债表.xlsx",index_col=0)  # 将Excel文件中的第0列数据作为数据集的index（行索引）
df
```

```python
# 2.保留“流动负债合计”“非流动负债合计”和“负债合计”这三行数据
df1 = df.loc[['流动负债合计','非流动负债合计','负债合计']]  # 按标签名获取数据
df1
```

```python
df1.index.name=None  # 取消行标签的名字
df1
```

```python
# 3. 根据下面的公式计算中国软件的流动负债和非流动负债在总负债中的占比，并将数据存入新列“占比”中。
import numpy as np
data1 = np.divide(df1.loc['流动负债合计'],df1.loc['负债合计'])  # 计算“流动负债占比”
data1 = data1[0]
print(data1)

data2 = np.divide(df1.loc['非流动负债合计'],df1.loc['负债合计'])  # 计算“非流动负债占比”
data2 = data2[0]
print(data1)
df1['占比']=[data1,data2,0]
df1
```

```python
# 删除无用行
# 参数axis为0表示删除行, inplace=True表示将结果更新到df1
df1.drop(['负债合计'],axis=0,inplace=True) 
df1
```

```python
# 4. 绘制饼图，展现中国软件的流动负债和非流动负债在总负债中的占比情况
import matplotlib.pyplot as plt  # 导入绘图包

plt.rcParams['font.sans-serif']=['SimHei']      # 设置正常显示中文
plt.rcParams['axes.unicode_minus']=False     # 设置正常显示负号

# 添加各种标签，使图形更直观
ratio = df1['占比'].tolist()  # 绘图数据，获取“占比”这一列的数据并转换成列表
ratio_text =df1.index    # df1的行标签作为扇形块标签文字
#colors = ["#e41a1c","#377eb8"]
plt.figure(figsize=(5,5))      # 指定figure的宽和高，单位为英寸
 
# 设置图形参数,绘制饼状图
plt.pie(
      ratio,                # 绘扇形块的数据
      labels=ratio_text,        # 扇形块标签文字
      autopct="%3.1f%%",    # 扇形块的数据标注格式，形如88.8%
      explode=(0,0.1),  # 第2个扇区凸出
      startangle=50,        # 饼图的初始摆放角度
      #colors=colors,        # 扇形块颜色列表
      )           
plt.title("中国软件负债占比分析")
```
