### （一）中执行下面的代码，并根据“利润表.xlsx”中的数据，解释各行代码的功能

```python
# 导入pandas库
import pandas as pd  
# 读入"利润表.xlsx"中工作表为 "2021年4月利润表 "的数据，保存为dfReport4对象
dfReport4 = pd.read_excel("利润表.xlsx",sheet_name="2021年4月利润表 ") 
dfReport4  # 显示dfReport4对象的内容
```

```python
dfReport4["日期"]='2021/04/30'  # 增加新列“日期”，值为'2021/04/30'
dfReport4
```

```python
# 重命名列标签
# 将"Unnamed: 1"重命名为"行次","Unnamed: 2"重命名为"本月数", "Unnamed: 3"重命名为 "本年累计数"，直接在原数据集上修改
dfReport4.rename(columns={"Unnamed: 1": "行次","Unnamed: 2": "本月数", "Unnamed: 3": "本年累计数"},inplace=True) 
dfReport4
```

```python

dfReport4.drop(['行次','本年累计数'], axis=1,inplace=True) # 删除'行次','本年累计数'这两列
dfReport4.drop(labels=[0,1,2],inplace=True)   # 删除第0,1,2行
dfReport4.fillna({"本月数":0},inplace=True)   # 将“本月数”列中的所有空值用0填充
dfReport4
```

## （二）完成读入5月和6月的利润表数据以及对这两个月数据进行清洗的任务

```python
dfReport5 = pd.read_excel("利润表.xlsx",sheet_name="2021年5月利润表 ")  # 读取Excel文件数据

# 添加一个日期列
dfReport5["日期"]='2021/05/31'

# 设置列标签
dfReport5.rename(columns={"Unnamed: 1": "行次","Unnamed: 2": "本月数", "Unnamed: 3": "本年累计数"},inplace=True)

# 删除无用的列
dfReport5.drop(['行次','本年累计数'], axis=1,inplace=True)

# 删除无用的行
dfReport5.drop(labels=[0,1,2],inplace=True)

# 将空值用0填充
dfReport5.fillna({"本月数":0},inplace=True)

dfReport5
```

```python
dfReport6 = pd.read_excel("利润表.xlsx",sheet_name="2021年6月利润表 ")  # 读取Excel文件数据

# 添加一个日期列
dfReport6["日期"]='2021/06/30'

# 设置列标签
dfReport6.rename(columns={"Unnamed: 1": "行次","Unnamed: 2": "本月数", "Unnamed: 3": "本年累计数"},inplace=True)

# 删除无用的列
dfReport6.drop(['行次','本年累计数'], axis=1,inplace=True)

# 删除无用的行
dfReport6.drop(labels=[0,1,2],inplace=True)

# 将空值用0填充
dfReport6.fillna({"本月数":0},inplace=True)

dfReport6
```

## （三）使用concat()对清洗后的4至6月数据进行纵向多行拼接合并

```python
# 纵向多行拼接合并
dfTotalRow=pd.concat([dfReport4,dfReport5,dfReport6])
dfTotalRow
```

## （四）执行下面的代码，说明这段代码计算的what是什么值。

```python
import numpy as np  # 导入numpy库
```

```python
# 查找'利润表'这列中值为'一、营业收入'这行，对应的“本月数”列的所有值,包含三行，4、5、6月三个月都有这个这
dfbak=dfTotalRow.loc[dfTotalRow['利润表']=='一、营业收入', '本月数']
dfbak
```

```python
# 计算，'利润表'这列中值为'一、营业收入'这行对应的“本月数”列的所有值的和
revenue = dfTotalRow.loc[dfTotalRow['利润表']=='一、营业收入', '本月数'].sum()  #得到4、5、6三个月的营业收入总和
revenue
```

```python
# 计算‘利润表'这列中值为'减：营业成本'这行，对应的“本月数”列的所有值的和
costs = dfTotalRow.loc[dfTotalRow['利润表']=='减：营业成本', '本月数'].sum()  #得到4、5、6三个月的营业成本总和
costs
```

```python
#4、5、6三个月的营业收入总和减去三个月的营业成本总和，除以三个月的营业收入总和
what=np.divide((revenue-costs),revenue)  # 计算销售毛利率
what
```
