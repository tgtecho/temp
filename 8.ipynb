### （一）对海淀区电信客户信息进行清洗
根据素材文件“电信客户流失数据-海淀区.csv”，完成对该文件的数据清洗工作，并将清洗好的数据另存为“result_haidian.csv”。
1. 读入数据文件“电信客户流失数据-海淀区.csv”，将数据保存在DataFrame对象hdDf中。
2. 检测数据集中有无缺失值，若整行都缺失，直接删除，并将删除结果更新至原数据集hdDf中。
3. 统计hdDf中哪些列有缺失值。
4. 分析有哪些列有缺失值，缺失值的数量有多少？
5. 为有缺失值的列设计合理的数据填充方案，并填充之。
6. 统计数据集中重复行的数量，对于完全重复的行，直接删除。
7. 观察每个列的重复值情况，判断这些重复是否符合常理。
8. 将清洗后的结果保存到外部文件“result_haidian.csv”中。

```python
# 1.
import pandas as pd
hdDf = pd.read_csv("电信客户流失数据-海淀区.csv")
hdDf.head()  # 显示前五行
```

```python
# 2.
# 将数据集中的空行直接删除
hdDf.dropna(how='all',inplace=True)
hdDf
```

```python
# 3.
print('hdDf总行数为：',len(hdDf))
for item in hdDf.columns:
    print('{}列是否有空值？'.format(item))
    if (hdDf[item].isnull().any()):
        print('有空值')
    else:
        print('无空值')
    print('-'*30)
```

```python
# 4.
number = len(hdDf)    # hdDf总行数
for item in hdDf.columns:
    print(item,":\n",hdDf[item].isnull().value_counts())
    if (hdDf[item].isnull().any()):
         # value_counts()[1]为空值数量，value_counts()[0]为非空值数量
        print('{}列有空值，空值数量为{}条'.format(item, hdDf[item].isnull().value_counts()[1])) 
    else:
        print('无空值')

    print('-'*30)
```

```python
# 5.
# 输出有空值的列名
for item in hdDf.columns:
    if (hdDf[item].isnull().any()):
        print('{}列有空值，空值数量为{}条'.format(item, hdDf[item].isnull().value_counts()[1])) 
        print('-'*30)
        
# 用平均通话时长填充“通话时长”列的空值
hdDf.fillna({"通话时长":hdDf["通话时长"].mean()},inplace=True)

# 对“上门服务”“无线时长”“影视时长”列用0填充空值
hdDf.fillna({"上门服务":0},inplace=True)
hdDf.fillna({"无线时长":0},inplace=True)
hdDf.fillna({"影视时长":0},inplace=True)

print(hdDf["通话时长"].isnull().any())
print(hdDf['上门服务'].isnull().any())
print(hdDf['无线时长'].isnull().any())
print(hdDf['影视时长'].isnull().any())
```

```python
# 6.
print('总行数',len(hdDf))    # hdDf总行数
# 统计重复行的数量
hdDf.duplicated().value_counts() 

# 若有空行，将数据集中的空行直接删除
# hdDf.dropna(how='all')
```

```python
# 修改设置，使系统显示所有列
pd.set_option('display.max_columns', None)
# 修改设置，使系统显示所有行
pd.set_option('display.max_rows', None)
```

```python
# 7.
# 统计“区域”列重复值的数量
print(hdDf.duplicated(subset=['区域']).value_counts())
```

```python
# 对于标记为True的数据，说明与其前面的值重复，观察区域列，索引为0的为False，其余都为True，说明所有值都是一样的
# 对于区域列来说，都应该是一样的海淀区
# 其他列请自行分析
for item in hdDf.columns:
    print(item,':\n',hdDf.duplicated(subset=[item]))
    print('-'*30)
```

```python
# 8.
hdDf.to_csv('result_haidian.csv',index=None,encoding='utf_8_sig')
```
